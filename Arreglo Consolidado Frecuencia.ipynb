```python
salida = []
referencia = list(df.columns[:3])
referencia.extend(list(df.columns[205:]))
for j in range(2005, 2013):
    refAuxiliar = referencia.copy()
    mesAnio = []
    for i in range(1, 13):
        mesAnio.append(f"{j}-{i}")
    refAuxiliar.extend(mesAnio)
    dfAnio = df[refAuxiliar]
    print (dfAnio.columns)
    for h in mesAnio :
        dfAnio[h[5:]] = dfAnio[h]
        del dfAnio[h]
    dfAnio["Año"] = j
    salida.append(dfAnio)
dfsalida = pd.concat(salida)
```
